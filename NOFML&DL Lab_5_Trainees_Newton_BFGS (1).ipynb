```python
def newton_method_optimization(x0,learning_rate ,n_iter=100, eps=0.1):
    x = x0
    i=0
    while i <(n_iter):
        
        if abs(first_derivate(x)) < eps:
            break
        x = x -(learning_rate) *(first_derivate(x) / second_derivate(x))
        i+=1
    print("x0 = ",x0)
    print("epsilon",eps)
    print("converged after",(i),"iterations")
    print("x_min = ",x)
    print("Gradient = ",first_derivate(x))
    
    
```

```python
def GD(x0,learning_rate ,n_iter=100, eps=0.1):
    x = x0
    i=0
    while i <(n_iter):
        
        if abs(first_derivate(x)) < eps:
            break
        x = x -(learning_rate) *(first_derivate(x))
        i+=1
    print("x0 = ",x0)
    print("epsilon",eps)
    print("lr",learning_rate)
    print("converged after",(i),"iterations")
    print("x_min = ",x)
    print("Gradient = ",first_derivate(x))
    
```

```python

def newton_optimization(x0, gradient, hessian, learning_rate, n_iter, eps):
    x = np.array(x0).reshape(-1, 1)
    
    i = 0
    
    while i < n_iter:
        grad = gradient(x[0,0],x[1,0])
        
        hess = hessian(x[0,0],x[1,0])
        
        if np.linalg.norm(grad) < eps:
            break
        if learning_rate == None:
            x =x-np.linalg.inv(hess) @ grad
        else:
            x =x- learning_rate * np.linalg.inv(hess) @ grad
        i += 1
    
    print("x0 =", x0)
    print("epsilon =", eps)
    print("converged after", i, "iterations")
    print("x_min =", x)
    print("Gradient =", grad)
```

```python

def GD(x0, gradient, learning_rate, n_iter, eps):
    x = np.array(x0).reshape(-1, 1)
    
    i = 0
    
    while i < n_iter:
        grad = gradient(x[0,0],x[1,0])
        
        if np.linalg.norm(grad) < eps:
            break
        
        x =x- learning_rate * grad
        i += 1
    
    print("x0 =", x0)
    print("epsilon =", eps)
    print("converged after", i, "iterations")
    print("x_min =", x)
    print("Gradient =", grad)
```
